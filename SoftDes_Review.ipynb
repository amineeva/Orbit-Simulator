Functions are chunks of code that be used multiple times in different contexts. Name, parameters, side effects, return value.

pass - statement that tells computer to pass over function content


```python
def reverse(word):
    """
    Return a reversed copy of a word.

    Args:
        word: A string representing the word to reverse.

    Returns:
        A string containing the characters of word in reverse order.
    """
    return word[::-1]

# common functions
print("Hello world")
len("hello world")
int(5.0) #this can be done with float(), bool(), str()
help(reverse)
```

Lists - sequence of items arranged in a specific order 

```python
# can be one type or mixed
digits = [3, 6, 7, 23, 6, 9]
mixed_list = [True, 890, "yuh", 4.20]

base_list = [1, 2, 3]
copy_list = [1, 2, 3]
other_list = [3, 2, 1]

print(base_list == copy_list)
print(base_list != copy_list)
print(base_list != other_list)

# can add and multiply list
print(base_list + copy_list)
print(base_list * 4)

# can uses indeces and slicing like in a string
digits[2]
digits[:3]
print(digits[::-1])

# can change elements via indices
print(digits)
digits[2] = 5
print(digits)

# check if in element
print(4 in digits)
print(9 in digits)

# list functions!!
print(sum(digits))
letters = ["ab", "cd", "ef"]
print("".join(letters))
print(min(digits))
print(max(digits))
print(sorted(digits)) # created a copy, this does NOT change digits

# methods
print(digits)
digits.sort()
print(digits)
digits.reverse()
print(digits)
print(digits.count(6)) #count of 6s in list
print(digits.index(5)) #index of first instance of 5
digits.append(17)
print(digits)
```

Loops: for and while

```python
# for loops - good if you know the number of iterations to go through
for seconds in list(range(10, 0, -1)):
    print(seconds)

# break - immediately exists loop
sentence = "Myself is me, I am me"
from string import punctuation
for character in sentence:
    if character in punctuation:
        print(character)
        break

# continue - skips rest of loop body 
for character in sentence:
    if character == " ":
        continue
    print(character)
```

Unit testing - test what the function actually does. The point is to test unique cases to see what's true and what isn't. Written like functions, need to start with 'test_', Softdes uses PyTest. Typically run in a separate testing file. 

```python
def average_value(numbers):
    """
    Return the average of a list of numbers.

    Args:
        numbers: A list of numbers (ints or floats)
    
    Returns:
        A float representing the average value of the numbers in the list.
    """
    return sum(numbers) / 3

def test_average_value_123(): # this one will pass because it's 3 numbers
    """
    Check that the average of the simple list [1, 2, 3] is 2.0
    """
    assert average_value([1, 2, 3]) == 2.0

def test_average_value_single_one():
    """
    Check that the average value of a single 1 is indeed 1.0.
    """
    assert average_value([1]) == 1.0
```
