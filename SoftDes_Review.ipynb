```python
# Integers and Arithmetic 
print(2 + 5)
print(21 - 34)
print(2 * 8)
print(2 ** 8)
print(5/2)
print(5//2) # floor operator, rounds down and returns int
print(5%2) # remainder operator, gives int remainder from floor 
print(13%-4) # can return negative remainders

# Bools can only be 'True' or 'False', capitalization matters here
```

Strings can be either "" or '', both are the same. Can use \" and \' to use ' and : in text. 

```python
print('There\'s no place like home')
# String operators
print("Hello" + " World")
print("ha" * 4)

# check if value in string
print("ah" in "ahahaha")
print("hello" in "Hello World")

# for printing with changing variables
name = "Alex"
print(f"Hello, {name}!")

# index starts at 0, the last value has an index of -1
school = 'Olin College of Engineering'
print(school[1])
print(school[-26])
print(school[-1])
#index a section of the string
print(school[0:5]) # inclusive 0, exclusive 5
print(school[5:]) # everything index 5 and after (inclusive)
print(school[:5]) #everything before index 5 (exclusive) 
#can skip numbers
print(school[::2]) #prints every other value
print(school[0::3])
print(school[::-1]) #prints backwards
```

If, elif, else

```python
birth_month = 1;
if birth_month == 1:
    written_month = "Jan"
elif birth_month == 2:
    written_month = "Feb"
else:
    written_month = "other"
print(written_month)

# apple item example
in_inventory = 1
item = "apple"

if in_inventory == 0:
    response = f"You have no {item}s."
elif in_inventory == 1:
    response = f"You have a {item}."
else:
    response = f"You have {in_inventory} {item}s."
print(response)
```
