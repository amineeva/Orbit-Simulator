Unit testing - test what the function actually does. The point is to test unique cases to see what's true and what isn't. Written like functions, need to start with 'test_', Softdes uses PyTest. Typically run in a separate testing file. 

```python
def average_value(numbers):
    """
    Return the average of a list of numbers.

    Args:
        numbers: A list of numbers (ints or floats)
    
    Returns:
        A float representing the average value of the numbers in the list.
    """
    return sum(numbers) / 3

def test_average_value_123(): # this one will pass because it's 3 numbers
    """
    Check that the average of the simple list [1, 2, 3] is 2.0
    """
    assert average_value([1, 2, 3]) == 2.0

def test_average_value_single_one():
    """
    Check that the average value of a single 1 is indeed 1.0.
    """
    assert average_value([1]) == 1.0
```

```python
# Example dictionary
number_words = {"one": 1, "two":2, "three": 3}
# one, two, three are KEYS
# 1, 2, 3 are VALUES

print(number_words["one"]) # calls the VALUE associated with KEY

# check if KEY 'taken'
print("zero" in number_words)
print("one" in number_words)

# add maps to dictionary:
number_words["zero"] = 0
print(number_words)

# change existing values:
number_words["two"] = 42
print(number_words)

# delete maps
del number_words["three"]
print(number_words)


# Loops
for VALUE in number_words:
    print(number_words[VALUE]) #calls the KEY, prints the VALUE

for KEY, VALUE, in number_words.items(): #allows printing both key and value
    print(f"The word {KEY} is the number {VALUE}")

squares = {i: i ** 2 for i in range(5)}
print(squares)


# Functions
print(len(number_words))
# can really only use for loops to iterate backwards/sorted:
for KEY in reversed(number_words):
    print(f"The word {KEY} is the number {number_words[KEY]}")
for KEY in sorted(number_words): # sorts alphabetically!
    print(f"The word {KEY} is the number {number_words[KEY]}")


# Methods
# iterate through values, not keys
for VALUE in number_words.values():
    print(f"number: {VALUE}")

# check if a key has a paired value (return 0 if not)
birthday = {"year": 1947, "month": 1, "day": 1}
birthday.get("year", 0)  # This returns 1947
birthday.get("hour", 0)  # This returns 0

# combine dictionaries with update:
current_time = {"year": 2025, "month": 3, "day": 24}
new_time = {"month": 8, "day": 30, "hour": 12, "minute": 30}
current_time.update(new_time)
print(current_time)
```

Recursion - break down problem into smaller versions until it can be easily solved, function calls itself. Want to break down smaller and smaller problems to get to the base case:

```python
def list_min(numbers):
    # if statement for when we hit base case
    if len(numbers) == 1:
        return numbers[0]
    else:
        # this is the recursive part
        first = numbers[0] # get head of list
        min_rest = list_min(numbers[1:]) #list_min calls itself here, finds min of rest of list

        # comparing head to rest of list
        if first < min_rest:
            return first
        else:
            return min_rest
```

good to user recursion under two conditions:

(1) you don't know how many iterations you need (for loop)

(2) it is not easy to track iterations (while loop)

```python
# Example - you need to buy asparagus but don't know what combination of bundles (6, 10, 15) to buy:
def asparagus(n):
    if n == 0:
        return True # buy nothing if you need 0 asparagus
    elif n < 6:
        return False # if less than 6, buy min bundle of 6
    else:
        return asparagus(n-6) or asparagus(n-10) or asparagus(n-15)
        # you're guaranteed to hit one of the base cases eventually
```
